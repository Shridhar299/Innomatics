```python
import re
import pandas as pd
import requests
import numpy as np
```

```python
path = r"F:\Innomatics\TASK-2\GMT20240125-121850_RecordingnewChat"
```

```python
import os
list1=[]
for i , j, k in os.walk(path):
    if len(k)>0:
        file = os.path.join(str(i)+ "\\" +str(k[0]))
        list1.append(file)
        
print(list1)
```

```python
finallist=[]
for i in list1:
    with open (r'{}'.format(i),"r" ,encoding = "utf-8") as txt:
        f = txt.readlines()
        finallist.append(f)
```

```python
finallist
```

```python
message=[]
for i in finallist:
    for j in i:
        
        m=re.findall(':\\t(\w.+)',j)
        if(m):
            message.append(m[0])
        else:
            message.append(np.nan)
```

```python
d1={}

d1['Message']=message
dm1=pd.DataFrame(d1)
dm1
```
